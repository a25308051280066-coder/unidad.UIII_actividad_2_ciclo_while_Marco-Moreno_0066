### Ejemplos de While

```python
meta = int(input("Ingresa la meta de ahorro: "))

ahorro = 0

while ahorro < meta:

    deposito = int(input("Ingresa la cantidad a ahorrar: "))

    if deposito > 0:
        ahorro = ahorro + deposito
        input("Ahorro actual: " + str(ahorro))
    else:
        input("Cantidad inválida")

input("Meta alcanzada")
```

```python
clave = ""
intentos = 0

while clave != "python123" and intentos < 3:

    clave = input("Ingresa la contraseña: ")

    if clave == "python123":
        input("Acceso permitido")
    else:
        intentos = intentos + 1
        input("Contraseña incorrecta")

if intentos == 3:
    input("Cuenta bloqueada")
```

```python
numero = 1
total = 0

while numero != 0:

    numero = int(input("Ingresa un número (0 para terminar): "))

    if numero != 0:
        total = total + numero
        input("Total acumulado: " + str(total))
    else:
        input("Fin del programa")

input("Resultado final: " + str(total))
```

```python
edad = 0

while edad < 18:

    edad = int(input("Ingresa tu edad: "))

    if edad >= 18:
        input("Acceso permitido")
    else:
        input("Eres menor de edad")
```

```python
presupuesto = int(input("Ingresa tu presupuesto: "))

gasto = 0

while gasto < presupuesto:

    compra = int(input("Ingresa el costo de la compra: "))

    if compra > 0 and gasto + compra <= presupuesto:

        gasto = gasto + compra

        input("Gasto actual: " + str(gasto))

    else:
        input("No puedes realizar esa compra")

input("Presupuesto agotado")
```
