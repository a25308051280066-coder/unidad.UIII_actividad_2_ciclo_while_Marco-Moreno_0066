#### Infografia de ciclo for

```python
# Ejemplo 1: recorrer números del 1 al 5

for i in range(1, 6):
    print(i)
```

```python
# Ejemplo 2: recorrer una lista

frutas = ["manzana", "pera", "uva", "kiwi"]

for fruta in frutas:
    print(fruta)
```

```python
# Ejemplo 3: suma de números

suma = 0

for i in range(1, 6):
    suma = suma + i

print("La suma es:", suma)
```

```python
# Ejemplo 4: recorrer una cadena

palabra = "Python"

for letra in palabra:
    print(letra)
```

```python
# Ejemplo 5: tabla de multiplicar del 3

tabla = 3

for i in range(1, 11):
    print(f"{tabla} x {i} = {tabla * i}")
```

```python
# Ejemplo 6: números pares del 1 al 10

for numero in range(1, 11):

    if numero % 2 == 0:
        print(numero)
```

```python
# Ejemplo 7: cuenta regresiva

for i in range(5, 0, -1):
    print(i)

print("Despegue")
```

```python
# Ejemplo 8: mensaje personalizado

nombre = "Marco"

for i in range(3):
    print("Hola", nombre)
```
